Binarize labels in a one-vs-all fashion

Several regression and binary classification algorithms are available in scikit-learn. A simple way to extend these algorithms to the multi-class classification case is to use the so-called one-vs-all scheme.

At learning time, this simply consists in learning one regressor or binary classifier per class. In doing so, one needs to convert multi-class labels to binary labels (belong or does not belong to the class). LabelBinarizer makes this process easy with the transform method.

At prediction time, one assigns the class for which the corresponding model gave the greatest confidence. LabelBinarizer makes this easy with the inverse_transform method.

## with thresholding

```python
data.describe()
```

```python
binarizer_elapsedDays = Binarizer(threshold=65)
binarizer_ratioDaysIden = Binarizer(threshold=0.19)
```

```python
elapsedDays = data['Elapsed Days In Sales Stage'].values.reshape(1, -1) 
ratioDaysIden = data['Ratio Days Identified To Total Days'].values.reshape(1, -1)
```

```python
elapsed_days_binary = binarizer_elapsedDays.fit_transform(elapsedDays)
ratio_days_identified_binary = binarizer_ratioDaysIden.fit_transform(ratioDaysIden)
```

```python
print ("\nOriginal elapsed days data values : \n",  elapsedDays) 
print ("\nOriginal 'Ratio Days Identified To Total Days' data values : \n",  ratioDaysIden) 
print ("\nBinarized elapsed days : \n", elapsed_days_binary) 
print ("\nBinarized 'Ratio Days Identified To Total Days' : \n", ratio_days_identified_binary) 
```
